```python
import pandas as pd
import re # Regex
```

```python
# Veriler "tutanak_party_csv" klasörünün içerisinde, hepsini ayrı ayrı çekiyorum

donem22 = pd.read_csv('tutanak_party_csv/22donem_all_party.csv')
donem23 = pd.read_csv('tutanak_party_csv/23donem_all_party.csv')
donem24 = pd.read_csv('tutanak_party_csv/24donem_all_party.csv')
donem25 = pd.read_csv('tutanak_party_csv/25donem_all_party.csv')
donem26 = pd.read_csv('tutanak_party_csv/26donem_all_party.csv')
donem27 = pd.read_csv('tutanak_party_csv/27donem_all_party.csv')
```

```python
df = pd.concat([donem22, donem23, donem24, donem25, donem26, donem27]) # verileri 'df' DataFrame'inde birleştiriyorum
df.head()
```

```python
words_pattern = r"[A-ZİÖÜĞÇŞa-zıöüğçş']\w+" # kelimeleri bölmek için regex pattern'ı

df['word_count'] = df['sentence'].apply(lambda x: len([word for word in re.findall(words_pattern, x)])) # 'sentence' sütununa regex uygulayarak 'word_count' sütunu oluşturuyor
```

```python
df['date'] = pd.to_datetime(df['date']) # date sütununu pandas'ın datetime objesine çeviriyor
df['year'] = df['date'].dt.year # datetime objesi içerisinden yıl bilgisini alarak yeni bir sütun olarak ekliyor
```

```python
df.head()
```

```python
df.groupby(['person', 'year'])['word_count'].sum() # isimlere ve yıllara göre gruplayarak kelime sayılarını topluyor
```

```python
pd.DataFrame(df.groupby(['person', 'year'])['word_count'].sum()).reset_index() # yukarıdakini DataFrame haline getiriyor
```

```python
word_count = pd.DataFrame(df.groupby(['person', 'year'])['word_count'].sum()).reset_index()
word_count.to_csv('word_count.csv', index=False) # Oluşturduğumuz DataFrame'i word_count.csv diye bir csv'ye export ediyor
```

```python
df.groupby(['person', 'year'])['sentence'].count() #kelime ve insanlara göre gruplayarak 'sentence'ların sayısına bakıyor
```

```python
pd.DataFrame(df.groupby(['person', 'year'])['sentence'].count()).reset_index() #yukarıdakini DataFrame haline getiriyor
```

```python
sentence_count = pd.DataFrame(df.groupby(['person', 'year'])['sentence'].count()).reset_index()
sentence_count.to_csv('sentence_count.csv', index=False) # Oluşturduğumuz DataFrame'i sentence_count.csv diye bir csv'ye export ediyor
```
