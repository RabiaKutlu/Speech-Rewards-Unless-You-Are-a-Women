# TBMM Tutanaklar Web Scraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## A scrapping function

```python
def url_to_text(url):
    r = requests.get(url)
    r.encoding = 'ISO 8859-9'
    unicode_str = r.content.decode('ISO 8859-9')
    encoded_str = unicode_str.encode("ISO 8859-9")
    soup = BeautifulSoup(encoded_str, 'html.parser')
    soup = soup.get_text().replace("ð", "ğ")
    return soup
```

```python
# Örnek url üzerinden deneme

print(url_to_text("https://www.tbmm.gov.tr/tutanak/donem27/yil3/ham/b11301h.htm"))
```

```python
# Text'in raw halinde (print fonksiyonu kullanmadan yazdırdığınızda) aşağıda replace ettiğim karakterlerden var bir sürü,
# o yüzden şöyle bir ön temizlemeyi de fonksiyona ekleyebilirim:
url_to_text("https://www.tbmm.gov.tr/tutanak/donem27/yil3/ham/b11301h.htm").replace('\n', '').replace('\xa0', '').replace('\r', ' ')
```

### Getting URLs

```python
tutanak_html = requests.get("https://www.bbc.co.uk/turkish/specials/1937_elections2007/page2.shtml")
tutanak_soup = BeautifulSoup(tutanak_html.text, 'html.parser')
```

```python
tutanak_urls = []
for link in tutanak_soup.find_all('a'):
    tutanak_urls.append(link.get('href'))
    
for url in tutanak_urls:
    print(url)
```

```python
### Veri setinde eksik olanlar
print("27. Dönem 3. Yasama Yılı:")
print(tutanak_urls[0])

print("27. Dönem 4. Yasama Yılı:")
print(tutanak_urls[1])
```

### 27. Dönem 3. Yasama Yılı

```python
donem27_yasama3_html = requests.get(tutanak_urls[1])
donem27_yasama3_soup = BeautifulSoup(donem27_yasama3_html.text, 'html.parser')
```

```python
donem27_yasama3_urls = []
for link in donem27_yasama3_soup.find_all('a'):
    donem27_yasama3_urls.append(link.get('href'))

for url in donem27_yasama3_urls:
    print(url)
```

```python
# "27. Dönem 3. Yasama Yılı için eksik olan tutanaklar
donem27_yasama3_urls[0:12:2]
```

### 27. Dönem 4. Yasama Yılı

```python
donem27_yasama4_html = requests.get(tutanak_urls[0])
donem27_yasama4_soup = BeautifulSoup(donem27_yasama4_html.text, 'html.parser')
```

```python
### Oylama ve Özetleri çıkardım

donem27_yasama4_urls = []
for link in donem27_yasama4_soup.find_all('a'):
    if "ozet" in link.get('href'):
        pass
    elif "oylama" in link.get('href'):
        pass
    else:
        donem27_yasama4_urls.append(link.get('href'))

for url in donem27_yasama4_urls:
    print(url)
```

## Creating a Dataset

```python
tutanak = pd.DataFrame(columns=['tutanak', 'dönem', 'yıl', 'birleşim'])
```

```python
for url in donem27_yasama3_urls[0:12:2]:
    data = {
        'tutanak': url_to_text(url),
        'dönem': 27,
        'yıl': 3,
        'birleşim': url[-10:-7]
    }
    tutanak = tutanak.append(data, ignore_index=True)
```

```python
for url in donem27_yasama4_urls:
    data = {
        'tutanak': url_to_text(url),
        'dönem': 27,
        'yıl': 4,
        'birleşim': url[-10:-7]
    }
    tutanak = tutanak.append(data, ignore_index=True)
```
